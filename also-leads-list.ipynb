```python
# join rslt_df with our marketing channel dataframe
rslt_df_final = rslt_df.merge(df_activecampaign, on='Email', how='left')

# rename the marketing channel columns
rslt_df_final = rslt_df_final.rename(columns={"*Marketing Channel": "Marketing Channel", "Account (previously Organization)":"Company Name"})
rslt_df_final.head()
```

```python
# drop everything except the columns that are needed
cols_needed = ['Date-Time',
               'Email',
               'First Name', 
               'Last Name', 
               'Phone Number',
               'Tag Name (For Reports)',
               'Company Name',
               'Country', 
               'Marketing Channel']

# create a final dataframe with only these columns
leads_list = rslt_df_final[cols_needed]
print(leads_list.info())

```

```python
# then remove if company name contains ALSO
leads_list = leads_list[~leads_list['Company Name'].str.contains("ALSO", case=False, na=False)]
```

```python
# loop for marketing channels wherever null - add web
leads_list['Marketing Channel'] = leads_list['Marketing Channel'].fillna(0)
for i in leads_list.index:
    if leads_list['Marketing Channel'][i] == 0:
        leads_list['Marketing Channel'][i] = 'Web'
    else:
        leads_list['Marketing Channel'][i] = leads_list['Marketing Channel'][i]

leads_list.head()

```
